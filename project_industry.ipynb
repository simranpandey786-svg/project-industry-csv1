### Descriptive Statistics for Numerical Columns

```python
import pandas as pd
pd.read_csv("INDUSTRY.csv")

# Load CSV file
df=pd.read_csv("INDUSTRY.csv")

# Select numeric columns
numeric_df = df.select_dtypes(include='number')

# Calculate statistics
mean = numeric_df.mean()
median = numeric_df.median()
mode = numeric_df.mode().iloc[0]
skewness = numeric_df.skew()
kurtosis = numeric_df.kurt()

# Display results
summary = pd.DataFrame({
    "Mean": mean,
    "Median": median,
    "Mode": mode,
    "Skewness": skewness,
    "Kurtosis": kurtosis
})

print(summary)
```

1. Descriptive Statistics Insights
Mean, Median, and Mode

The mean and median values for most variables are very close to each other, indicating that the dataset is well-balanced and does not suffer from significant distortion caused by extreme values. The mode values show the most frequently occurring observations, although the dataset is highly diverse and does not exhibit strong clustering around a single value.Skewness

The skewness values for all numerical variables are very close to zero. This indicates that the distributions are approximately symmetric, meaning the data is evenly spread on both sides of the average. There is no strong positive or negative skewness in any variable.

Kurtosis

All variables have kurtosis values close to -1.2, indicating a platykurtic distribution. This means the distributions are flatter than a normal distribution and contain fewer extreme values or outliers. The data appears relatively stable and evenly distributed.


### Visualization of Descriptive Statistics

```python
fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(12, 25))
fig.suptitle('Descriptive Statistics of Numerical Columns', fontsize=16, y=1.02)

stats_to_plot = ['Mean', 'Median', 'Mode', 'Skewness', 'Kurtosis']

for i, stat in enumerate(stats_to_plot):
    sns.barplot(x=summary.index, y=summary[stat], ax=axes[i], palette='viridis', hue=summary.index, legend=False)
    axes[i].set_title(f'{stat} of Numerical Columns')
    axes[i].set_ylabel(stat)
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### Comparison of Mean and Median for Numerical Columns

```python
mean_median_df = summary[['Mean', 'Median']].reset_index()
mean_median_df = mean_median_df.melt(id_vars='index', var_name='Statistic', value_name='Value')

plt.figure(figsize=(14, 7))
sns.barplot(x='index', y='Value', hue='Statistic', data=mean_median_df, palette='deep')
plt.title('Comparison of Mean and Median for Numerical Columns')
plt.xlabel('Numerical Column')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Statistic')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Difference Between Mean and Median

```python
mean_median_difference = summary['Mean'] - summary['Median']
print("Difference between Mean and Median for each numerical column:")
print(mean_median_difference)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("INDUSTRY.csv")

# Select numeric columns
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"\nColumn: {col}")
    print("Number of Outliers:", len(outliers))
```

#The dataset shows minimal evidence of extreme outliers. This is supported by:

Skewness values close to zero.
Negative kurtosis values.
Balanced mean and median values.

The absence of major outliers improves the reliability of statistical analysis and predictive modeling.





```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("INDUSTRY.csv")

numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='industry', y='market_rating', data=df, palette='viridis', hue='industry', legend=False)
plt.title('Market Rating Across Industry Sectors')
plt.xlabel('Industry')
plt.ylabel('Market Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
average_market_rating_by_region = df.groupby('region')['market_rating'].mean().reset_index()
print("Average Market Rating by Region:")
print(average_market_rating_by_region)
```

```python
plt.figure(figsize=(12, 6))
mean_profit_margin_by_industry = df.groupby('industry')['profit_margin_percent'].mean().sort_values(ascending=False)
sns.barplot(x=mean_profit_margin_by_industry.index, y=mean_profit_margin_by_industry.values, palette='viridis', hue=mean_profit_margin_by_industry.index, legend=False)
plt.title('Mean Profit Margin by Industry')
plt.xlabel('Industry')
plt.ylabel('Mean Profit Margin (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='region', y='market_rating', data=average_market_rating_by_region, palette='coolwarm', hue='region', legend=False)
plt.title('Average Market Rating by Region')
plt.xlabel('Region')
plt.ylabel('Average Market Rating')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
